```python
from twin.pipelines import graph_embedding_all
from twin.plots import drawedges
import twin.embedding as emb
import twin.dataset as dataset

import matplotlib.pyplot as plt
```

```python
def plot_embedding(G, embedding, embedding_kw={}):
    experiment = graph_embedding_all(
        G,
        embedding=embedding,
        embedding_kw=embedding_kw,
    )
    
    fig, axes = plt.subplots(1, 3, figsize=(3 * 6, 1 * 6), squeeze=True)
    
    for i, method in enumerate(["VertexEmbedding", "EdgeEmbedding", "TwinEmbedding"]):
        ax = axes[i]
    
        ax.set_title(method)
        ax.scatter(*experiment[method]["V"], c="r", s=18, zorder=3)
        drawedges(G, 2, experiment[method]["V"], ax, 0.3, 1)
    
    fig.suptitle(embedding.method_name, fontweight='bold')
    
    fig.tight_layout()
```

```python
G = dataset.football_2023()
```

```python
plot_embedding(G, emb.SpringLayout)
```

```python
plot_embedding(G, emb.YifanHuLayout)
```

```python
plot_embedding(G, emb.ForceAtlas2Layout)
```

```python
plot_embedding(G, emb.ForceAtlas2Layout, {'linlog': True})
```

```python
plot_embedding(G, emb.KamadaKawaiLayout, {'mode': 'similarity'})
```

```python
plot_embedding(G, emb.NeatoLayout, {'mode': 'similarity'})
```

```python
plot_embedding(G, emb.SpectralLayout)
```

```python
plot_embedding(G, emb.SGtSNELayout, {'lambda_par': 1.0})
```
