```python
from twin import dataset, metadata
from twin import embedding as emb
from twin.pipelines import embedding_metrics_all, metrics_to_df
from twin.config import REAL_WORLD_DIR

import numpy as np
import pandas as pd
import scipy.io as sio
```

```python
football_graph = dataset.football_2023()
football_standing = metadata.football_2023['conference_standing']

label_map = {standing: idx for idx, standing in enumerate(set(football_standing))}
vertex_labels = [label_map[standing] for standing in football_standing]
```

```python
football_metrics = embedding_metrics_all(football_graph, emb.SGtSNELayout, target=vertex_labels, n_samples=10, n_neighbors=[5,10,15,20])
```

```python
football_dfs = metrics_to_df(football_metrics)
```

```python
football_dfs['PointDistance']
```

```python
football_dfs['KNeighborsPreservation']
```

```python
football_dfs['KNeighborsClassifier']
```

```python
mnist_graph = dataset.mnist(10)
mnist_digits = metadata.mnist['digit']
```

```python
mnist_metrics = embedding_metrics_all(mnist_graph, emb.SGtSNELayout, target=mnist_digits, n_samples=10, n_neighbors=[100,200,300,400])
```

```python
mnist_dfs = metrics_to_df(mnist_metrics)
```

```python
tabula_graph = dataset.tabula('liver')

s = sio.loadmat(str(REAL_WORLD_DIR / 'tabula-liver.mat'))
orig_labels = s['cell_ontology_class'][0]
extracted_labels = np.array([elem[0] for elem in orig_labels])

label_map = {label: idx for idx, label in enumerate(set(extracted_labels))}
vertex_labels = [label_map[label] for label in extracted_labels]
```

```python
tabula_metrics = embedding_metrics_all(tabula_graph, emb.SGtSNELayout, target=vertex_labels, n_samples=10, n_neighbors=[100,200,300,400])
```

```python
tabula_dfs = metrics_to_df(tabula_metrics)
```

```python
keys = ['MNIST handwritten digits', 'Tabula Sapiens Liver']
dfs = [mnist_dfs, tabula_dfs]
```

```python
pd.concat([df['PointDistance'] for df in dfs], keys=keys, names=['Dataset'])
```

```python
pd.concat([df['KNeighborsPreservation'] for df in dfs], keys=keys, names=['Dataset'])
```

```python
pd.concat([df['KNeighborsClassifier'] for df in dfs], keys=keys, names=['Dataset'])
```
